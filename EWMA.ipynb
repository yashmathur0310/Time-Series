```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
airline=pd.read_csv('airline_passengers.csv',index_col=0,parse_dates=True)
airline
```

```python
airline.dropna(inplace=True)
```

```python
airline.index=pd.to_datetime(airline.index)
```

```python
airline['6 month moving average ']=airline['Thousands of Passengers'].rolling(6).mean()
```

```python
airline['12 month moving average ']=airline['Thousands of Passengers'].rolling(12).mean()
```

```python

airline.plot(figsize=(10,8))
```

```python
airline['EWMA-12']=airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['Thousands of Passengers','EWMA-12']].plot(figsize=(10,8))
```
